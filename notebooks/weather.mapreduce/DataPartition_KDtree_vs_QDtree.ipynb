```python
def draw_screen_poly(plt, gid, west, east, south, north, m, color, text=True):
    def fix(x):
        if x >  80: return 80.0
        if x < -80: return -80.0
        return x    
    lons = [west, east, east, west]
    lats = map(fix, [south, south, north, north])
    x, y = m( lons, lats )
    xy = zip(x,y)    
    poly = Polygon( xy, facecolor=color, alpha=0.4, label = str(gid) )

    if text:
        # center    
        clon = (west + east) / 2
        clat = fix((north + south) / 2)
        cx,cy = m([clon],[clat])
        plt.text(cx[0], cy[0], str(gid), 
                 horizontalalignment='center', 
                 verticalalignment='center',
                 bbox=dict(facecolor='white', alpha=1),
                 size =15, fontweight='bold')  
       
    plt.gca().add_patch(poly)   
```

```python
# This function defines the KD partition process

from __future__ import division
import pandas as pd
from mrjob.job import MRJob
import random

#count will recode the number of valid measurement
count = 0
class Node:
    #elements of node class
    def __init__(self,la,lo,il,n,l,r,ii=[]):
        #latitude
        self.latitude  = la
        #logitude
        self.longitude = lo
        # leaf indicator
        self.isleaf = il
        # number of valid measurements
        self.num = n
        # leftchild
        self.left = l
        # rightchild
        self.right = r
        #station index
        self.item = ii
    
    #print this node
    def __str__(self):
        print 'Is Leaf: ', self.isleaf
        print 'latitude:', self.latitude
        print 'longitude', self.longitude
        if self.left != None: print 'left\n', self.left
        if self.right != None: print 'right\n', self.right

class KDtree:
    #constructor of KD tree: (1)"bincap" is the threshod of  valid measurement of each node. If the number of valid 
    # measuremetn in this node is less than this threshold, stop partition. Otherwise, recursively partition the tree
    #(2)"stations" represents the stations information sent to the KD tree . (3) "rand_num" contains available labels for leaves of the KDtree
    def __init__(self, bincap, stations,rand_num):
        #
        global count
        count = 0
        #region index
        self.regions = rand_num
        #threshold
        self.bincap = bincap
        #station data
        self.stations = stations
        #tree root
        self.root = self.creat_tree(stations)
    
    # depth is the depth of this KDtree
    def creat_tree(self, stations, depth=0):
        global count 
        cum_sum = stations.weight.cumsum()
        #if valid measurement number is less than the threshold,
        #this node is a leaf. All the stations in this node are appened with
        #a certain region index  which is sampled from availabel labels "region" by "count"
        if cum_sum.iloc[-1] <= self.bincap:
            print '-----Leaf-----', cum_sum.iloc[-1], '----Stations----', len(stations)
            count += 1
            self.stations.loc[stations.index,'label']= self.regions[count]
            return Node(None,None,True,self.regions[count],None,None,stations.index)
                      
        isLa = False
        isLo = False
        #if depth is even, then we sort along the latitude
        if depth % 2 == 0:
            isLa = True
            stations.sort('latitude',ascending=True,inplace=True)
        #if depth is odd, then we sort along the longitude
        else:
            isLo = True
            stations.sort('longitude',ascending=True,inplace=True)
        #Median represents the partition line. 
        #The valid measurements number on the left side of the partition line is equal to the right
        median = 0 
        cum_sum = stations.weight.cumsum()
        for i in range(len(cum_sum)):
            if cum_sum.iloc[i] >= cum_sum.iloc[-1] / 2:
                median = i 
                break
        #if more than half valid measurements are in the last station, 
        #we seperate the last station from the rest
        if median == len(cum_sum)-1:
            median -= 1
        
        #We recursively partition the data along the latitude
        if isLa == True:    
            return Node(stations.iloc[median]['latitude']-0.00001,
                        None, False, None,
                        self.creat_tree(stations[:median+1],depth+1),
                        self.creat_tree(stations[median+1:],depth+1))
        #We recursively partition the data along the longitude
        elif isLo == True:
            return Node(stations.iloc[median]['longitude']-0.00001,
                        None, False, None,
                        self.creat_tree(stations[:median+1],depth+1),
                        self.creat_tree(stations[median+1:],depth+1))
        
```
