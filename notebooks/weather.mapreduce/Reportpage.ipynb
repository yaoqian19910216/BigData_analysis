** (2) Now we will instantiate the merging process. Whether or not two nodes can be merged is determined by the rule of explanation length. Assume we have two child nodes A and B, and their parent node C. $n_A$,$n_B$,and $n_C$ represent the number of data sets in node A, B and C, respectivly.  $k_A$, $k_B$,and $k_C$ represents the number of eigenvectors of each node after implementing the PCA analysis. if we have**

$$n_A * k_A + (k_A+1)\cdot(2*365) + n_B * k_B + (k_B+1)\cdot(2*365) > n_{C} * k_{C} + (k_{C}+1)\cdot(2*365)$$
**,we can merge two node. Intuitively speaking, two nodes can be merged together if the two regions have similar climate. We will further show that, if any adjacent subregions can be merged, our code is able to merge all of them.**

**In order to implement PCA on the data sets of each region, we seperate the whole data into different sub-files according to their region index. Those files are used in PCA step. **

>** PCA map-reducer **

> mapper1: [_, value = data] ====>[key = [station,year],values = ['TMAX'/'TMIN', weather_data] ]
>> Output the data with 'TMAX'/'TMIN'

> reducer1: [key = [station,year],values = vector(['TMAX'/'TMIN', weather_data]) ]  ====> [key = regionindex, value = [weather_max, weather_min] ]
>>Combine the TMAX anf TMIN data for one station in the same year

>>Check whether it is a valid measurement, which defined in the last section
    
> reducer2: [key = regionindex, value = [weather_max, weather_min] ] ====> [key = region, value = explanation length]
>> do PCA for all weather data in one region

>> return the explanation length for 95% / 99%



** The optimal algorithm is to traverse the whole graph. However, this will take very very very long time!!!! Instead, We use the random merging algorithm.**
> First, we will randomly pick up a point from the map and then merge all its possible neighbors. 

> Secondly, the previously merged node will merge to its new neighbors. By setting the merger steps, the merged area will gradually expand as long as it can be merged with any of its neighbors.

> We repeatedly choose 50 random starting points such that large part of mergable regions are merged together on the map.
