```python
from __future__ import division
import pandas as pd
import random

count = 0
class QuadTree:
    def __init__(self, stations, depth, bounding_rect, maxIt):
        global count
        self.nw = self.ne = self.se = self.sw = None
        self.maxIt = maxIt
        self.depth = depth
        self.bounding_rect = bounding_rect
        cum_sum = stations.weight.cumsum()
        if cum_sum.iloc[-1] <= self.maxIt and depth > 3:
            print '-----Leaf-----', cum_sum.iloc[-1], '----Stations----', len(stations)
            ML_station.loc[stations.index,'label']= count
            #DD.loc[stations.index,'label']= regions[self.depth]
            self.region = count
            count += 1
            self.items = stations.index
            self.leaf = True
            return
        self.leaf = False
        
        l, t, r, b = bounding_rect
        cx = self.cx = ( l + r ) * 0.5 
        cy = self.cy = ( t + b ) * 0.5 

        self.items = []
        nw_items = []
        ne_items = []
        se_items = []
        sw_items = []

        for item in stations.index:
            item_x = stations.loc[item, 'latitude']
            item_y = stations.loc[item, 'longitude']
            
            in_nw = item_x <= cx and item_y >  cy  
            in_ne = item_x >  cx and item_y >= cy
            in_sw = item_x <= cx and item_y <  cy  
            in_se = item_x >  cx and item_y <= cy
            
            if   in_nw: nw_items.append(item)
            elif in_ne: ne_items.append(item)
            elif in_sw: sw_items.append(item)
            elif in_se: se_items.append(item)

        if nw_items:
            self.nw = QuadTree(stations.loc[nw_items], depth+1, (l, t, cx, cy),maxIt)
        if ne_items:
            self.ne = QuadTree(stations.loc[ne_items], depth+1, (cx, t, r, cy),maxIt)
        if se_items:
            self.se = QuadTree(stations.loc[se_items], depth+1, (cx, cy, r, b),maxIt)
        if sw_items:
            self.sw = QuadTree(stations.loc[sw_items], depth+1, (l, cy, cx, b),maxIt)
```

##### Node Structure for Graph Model

```python
from sets import Set
class node:
    def __init__(self, r, br, b):
        self.region_index = r
        self.ex_length = 0
        self.rect = []
        self.rect.append(br)
        self.area = b
        self.neighbours = Set([])
        
    def set_length(self, l):
        self.ex_length = l
        
    def add_data(self, new_index):
        self.data_index.append(new_index)
    
    def add_neighbour(self, new_n):
        self.neighbours = self.neighbours.union(new_n)
    
    def del_neighbour(self, del_n):
        self.neighbours.discard(del_n)
        
```

```python
def leafs_to_node_list(root, node_list, region_hash):
    if root.leaf == True:
        [l, t, r, b] = root.bounding_rect
        new_node = node(root.region, root.bounding_rect, abs(l-r))
        region_hash[root.region] = new_node
        node_list.append(new_node)
    else:
        if root.nw != None: leafs_to_node_list(root.nw, node_list, region_hash)
        if root.ne != None: leafs_to_node_list(root.ne, node_list, region_hash)
        if root.sw != None: leafs_to_node_list(root.sw, node_list, region_hash)
        if root.se != None: leafs_to_node_list(root.se, node_list, region_hash)
```

```python
def is_adjacent(region1, region2):
    [b_1, r_1, t_1, l_1] = region1
    [b_2, r_2, t_2, l_2] = region2
    
    if l_1 == l_2 or r_1 == r_2:
        if t_1 == b_2 or b_1 == t_2:
            return True
        else:
            return False        
    if t_1 == t_2 or b_1 == b_2:
        if r_1 == l_2 or l_1 == r_2:
            return True
        else:
            return False
    if r_1 == l_2 or l_1 == r_2:
        if t_1 >= t_2 and b_1 <= b_2:
            return True
        elif t_2 >= t_1 and b_2 <= b_1:
            return True
        else:
            return False
    if t_1 == b_2 or b_1 == t_2:
        if l_1 >= l_2 and r_1 <= r_2:
            return True
        elif l_2 >= l_1 and r_2 <= r_1:
            return True
        else:
            return False
    
    return False
        
```

```python
def draw_screen_poly(plt, gid, west, east, south, north, m, color, text=True):
    def fix(x):
        if x >  80: return 80.0
        if x < -80: return -80.0
        return x    
    lons = [west, east, east, west]
    lats = map(fix, [south, south, north, north])
    x, y = m( lons, lats )
    xy = zip(x,y)    
    poly = Polygon( xy, facecolor=color, alpha=0.4, label = str(gid) )

    if text:
        # center    
        clon = (west + east) / 2
        clat = fix((north + south) / 2)
        cx,cy = m([clon],[clat])
        plt.text(cx[0], cy[0], str(gid), 
                 horizontalalignment='center', 
                 verticalalignment='center',
                 bbox=dict(facecolor='white', alpha=0.5),
                 size =9)  
       
    plt.gca().add_patch(poly)   
```

```python
import random
class station_graph:
    # initial the Graph with nodes
    def __init__(self, node_list):
        self.nodes = node_list
        self.node_num = len(node_list)
        self.merged = -1 * np.ones([len(node_list),])
    
    # create the Graph, add a link between two neighbour
    def create_graph(self):
        for i in range(self.node_num):
            for j in range(i+1, self.node_num):
                if is_adjacent(self.nodes[i].rect[0], self.nodes[j].rect[0]):
                    self.nodes[i].neighbours.add(self.nodes[j])
                    self.nodes[j].neighbours.add(self.nodes[i])
    
    def print_neighbour(self, n):
        for nn in self.nodes[n].neighbours:
            for j in range(self.node_num):
                if self.nodes[j] == nn:
                    print j, ' ',
                    
    # merge n2 to n1 and delete n2
    def merge_node(self, n1, n2):
        if self.merged[n1] != -1:
            self.merge_node(int(self.merged[n1]), n2)
            return
        if self.merged[n2] != -1:
            self.merge_node(n1, int(self.merged[n2]))
        
        if True:
            self.nodes[n2].neighbours.remove(self.nodes[n1])
            self.nodes[n1].neighbours = \
                self.nodes[n1].neighbours.union(self.nodes[n2].neighbours)
            self.nodes[n1].rect += self.nodes[n2].rect
            self.nodes[n1].neighbours.remove(self.nodes[n2])
            
            for nn in self.nodes[n2].neighbours:
                nn.neighbours.add(self.nodes[n1])
                nn.neighbours.discard(self.nodes[n2])
            self.merged[n2] = n1       
    def merge_neighbour(self, n):
        for nn in self.nodes[n].neighbours:
            ind = 0
            for j in range(self.node_num):
                if self.nodes[j] == nn:
                    self.merge_node(n, j)
                    break
                               
    # generate a random color
    def random_color(self):
        r = lambda: random.randint(0,255)
        return '#%02X%02X%02X' % (r(),r(),r())
    
    # draw the regions with different color
    def draw_regions(self, isNum=False, li=-130, la=-60,ti=20, ta=50):
        lonmin=li;lonmax=la;latsmin=ti;latsmax=ta;
        plt.figure(figsize=(15,10),dpi=300)
        m = Basemap(projection='merc',llcrnrlat=latsmin,urcrnrlat=latsmax,\
            llcrnrlon=lonmin,urcrnrlon=lonmax,lat_ts=20,resolution='i') #'c'
        m.drawcoastlines()

        for i in range(self.node_num):
            if self.merged[i] != -1: continue
            color = self.random_color()
            for rect in self.nodes[i].rect:
                [b, r, t, l] = rect
                if isNum == False:draw_screen_poly(plt, '', l, r, t, b, m, color)
                else: draw_screen_poly(plt, str(i), l, r, t, b, m, color)
                
        plt.title('weather stations partitioning')
        plt.show()
    
    # draw all neighbours for region k
    def draw_one_neighbour(self, k):
        if self.merged[k] != -1:
            self.draw_one_neighbour(int(self.merged[k]))
            return 
        
        lonmin=-130;lonmax=-60;latsmin=20;latsmax=50;
        plt.figure(figsize=(15,10),dpi=300)
        m = Basemap(projection='merc',llcrnrlat=latsmin,urcrnrlat=latsmax,\
            llcrnrlon=lonmin,urcrnrlon=lonmax,lat_ts=20,resolution='i') #'c'
        m.drawcoastlines()
        # all region in blue
        for i in range(self.node_num):
            if self.merged[i] != -1: continue
            for rect in self.nodes[i].rect:
                [b, r, t, l] = rect  
                draw_screen_poly(plt, '', l, r, t, b, m, '#58D3F7') 
        # k region to be red
        for rect in self.nodes[k].rect:
            b, r, t, l = rect
            draw_screen_poly(plt, '', l, r, t, b, m, '#FE2E2E')
        # neighbour to be yellow
        for nn in self.nodes[k].neighbours:
            for rect in nn.rect:
                b, r, t, l = rect
                draw_screen_poly(plt, '', l, r, t, b, m, '#FFBF00')
                
        plt.title('weather stations partitioning')
        plt.show()
```

```python
def merge_pca_weather_node(Graph_ss, ind_list):
    
    data = []
    for ii in ind_list:
        r1 = Graph_ss.nodes[ii].region_index
        filename1 = '/home/ubuntu/station_data/' + str(r1)+'.0_data.pklz'
        f1 = gzip.open(filename1, 'rb')
        pickleFile1 = cPickle.Unpickler(f1)
        data1 = pickleFile1.load()
        data += data1
        f1.close()
    
    M = pd.DataFrame(data)
    M[M == ''] = float('NaN')
    M = M.astype(float)
    M = M.transpose()
    (columns,rows)= np.shape(M)
    Mean = np.mean(M, axis=1).values
    C=np.zeros([columns,columns])  
    N=np.zeros([columns,columns])
    
    for i in range(rows):
        row = M.iloc[:,i] - Mean
        outer = np.outer(row,row)
        valid = np.isnan(outer) == False
        C[valid] = C[valid]+ outer[valid]
        N[valid] = N[valid]+ 1
            
    valid_outer = np.multiply(1 - np.isnan(N),N>0)
    cov = np.divide(C,N)
    cov = np.multiply(cov, valid_outer)
    U, D, V = np.linalg.svd(cov)
    cum_sum = np.cumsum(D[:])/np.sum(D)
    for i in range(len(cum_sum)):
        if cum_sum[i] >= 0.99:
            ind = i 
            break
    f = gzip.open('cummmm.pklz','wb')
    pickle.dump(cum_sum,f)
    f.close()
    return ind, cum_sum
```

##### Show that 19 and 18 can not merge

For two node, A and B, on the same level, for their parent node C, if we have 
$$n_A * k_A + (k_A+1)\cdot(2*365) + n_B * k_B + (k_B+1)\cdot(2*365) > n_{C} * k_{C} + (k_{C}+1)\cdot(2*365)$$
we can merge two node.

If child node A have two region in it, we will skip that level and search other possible merge.

```python
def total_variable_num(data_num, exp_len):
    return data_num * exp_len + (exp_len+1)*(2*365)
```

```python
def merge_pca_weather(node1, node2):
    r1 = node1.region_index
    r2 = node2.region_index
    filename1 = '/home/ubuntu/station_data/' + str(r1)+'.0_data.pklz'
    filename2 = '/home/ubuntu/station_data/' + str(r2)+'.0_data.pklz'
    f1 = gzip.open(filename1, 'rb')
    pickleFile1 = cPickle.Unpickler(f1)
    data1 = pickleFile1.load()
    f2 = gzip.open(filename2, 'rb')
    pickleFile2 = cPickle.Unpickler(f2)
    data2= pickleFile2.load()
    f1.close()
    f2.close()
    
    len1 = shape(data1)[0]
    len2 = shape(data2)[0]
       
    data = data1 + data2
    
    M = pd.DataFrame(data)
    M[M == ''] = float('NaN')
    M = M.astype(float)
    M = M.transpose()
    (columns,rows)= np.shape(M)
    Mean = np.mean(M, axis=1).values
    C=np.zeros([columns,columns])  
    N=np.zeros([columns,columns])
    
    for i in range(rows):
        row = M.iloc[:,i] - Mean
        outer = np.outer(row,row)
        valid = np.isnan(outer) == False
        C[valid] = C[valid]+ outer[valid]
        N[valid] = N[valid]+ 1
            
    valid_outer = np.multiply(1 - np.isnan(N),N>0)
    cov = np.divide(C,N)
    cov = np.multiply(cov, valid_outer)
    U, D, V = np.linalg.svd(cov)
    cum_sum = np.cumsum(D[:])/np.sum(D)
    for i in range(len(cum_sum)):
        if cum_sum[i] >= 0.99:
            ind = i 
            break

    rule1 = total_variable_num(len1, node1.ex_length)
    rule2 = total_variable_num(len2, node2.ex_length)
    new_rule = total_variable_num(len1+len2, ind)
    
    isMerge = ( new_rule < rule1 + rule2 )
    
    return isMerge, ind, cum_sum
```

```python
def check_and_merge(Graph_ss, region_hash):
    n1 = randint(0,Graph_ss.node_num)
    while Graph_ss.merge_node[n1] != -1:
        n1 = randint(0,Graph_ss.node_num)
    r1 = Graph_ss.nodes[n1].region_index
    min_area = 360
    n2 = -1
    for nn in region_hash[r1].neighbours:
        if nn.area < min_area:
            n2 = nn.region_index
            min_area = nn.area
    
    [isMerge, ind] = merge_pca_weather(region_hash[n1], region_hash[n2])
    if isMerge == True:
        Graph_ss.merge_node(n1, n2)
        Graph_ss.draw_one_neighbour(n1)
        region_hash[n1].ex_length = ind       
```
