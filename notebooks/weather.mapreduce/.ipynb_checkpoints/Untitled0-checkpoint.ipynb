```python
%%writefile valid_measurement.py
from __future__ import division
from mrjob.job import MRJob

class ValidMeasurement(MRJob):
    def mapper1(self, _, line):
        elements = line.split(',')
        if elements[1] == 'TMIN' or elements[1] == 'TMAX':
            yield (elements[0], elements[2]), elements[3:]

    def reducer1(self, pair, vectors):
        vectors = list(vectors)
        if len(vectors) < 2:
            yield pair[0], 0
        else:
            day_count = 0 
            for i in range(len(vectors[0])):
                if vectors[0][i] != '' and vectors[1][i] != '': 
                    day_count += 1
            if day_count / len(vectors[0]) < 0.5:
                yield pair[0], 0
            else:
                yield pair[0], 1
    
    def reducer2(self, station, is_valid):
        weight = sum(is_valid)
        if weight != 0:
            yield station, weight
    
    def steps(self):
        return [self.mr(mapper=self.mapper1, reducer=self.reducer1),
                self.mr(reducer=self.reducer2)]

if __name__ == '__main__':
    ValidMeasurement.run()
```

```python
def draw_screen_poly(plt, gid, west, east, south, north, m, color, text=True):
    def fix(x):
        if x >  80: return 80.0
        if x < -80: return -80.0
        return x    
    lons = [west, east, east, west]
    lats = map(fix, [south, south, north, north])
    x, y = m( lons, lats )
    xy = zip(x,y)    
    poly = Polygon( xy, facecolor=color, alpha=0.4, label = str(gid) )

    if text:
        # center    
        clon = (west + east) / 2
        clat = fix((north + south) / 2)
        cx,cy = m([clon],[clat])
        plt.text(cx[0], cy[0], str(gid), 
                 horizontalalignment='center', 
                 verticalalignment='center',
                 bbox=dict(facecolor='white', alpha=1),
                 size =15, fontweight='bold')  
       
    plt.gca().add_patch(poly)   
```

```python
from __future__ import division
import pandas as pd
from mrjob.job import MRJob
import random

count = 0
class Node:
    def __init__(self,la,lo,il,n,l,r,ii=[]):
        self.latitude  = la
        self.longitude = lo
        self.isleaf = il
        self.num = n
        self.left = l
        self.right = r
        self.item = ii
    def __str__(self):
        print 'Is Leaf: ', self.isleaf
        print 'latitude:', self.latitude
        print 'longitude', self.longitude
        if self.left != None: print 'left\n', self.left
        if self.right != None: print 'right\n', self.right

class KDtree:
    def __init__(self, bincap, stations,rand_num):
        global count
        count = 0
        self.regions = rand_num
        self.bincap = bincap
        self.stations = stations
        self.root = self.creat_tree(stations)
    
    def creat_tree(self, stations, depth=0):
        global count 
        cum_sum = stations.weight.cumsum()
        if cum_sum.iloc[-1] <= self.bincap:
            print '-----Leaf-----', cum_sum.iloc[-1], '----Stations----', len(stations)
            count += 1
            self.stations.loc[stations.index,'label']= self.regions[count]
            return Node(None,None,True,self.regions[count],None,None,stations.index)
                      
        isLa = False
        isLo = False
        if depth % 2 == 0:
            isLa = True
            stations.sort('latitude',ascending=True,inplace=True)
        else:
            isLo = True
            stations.sort('longitude',ascending=True,inplace=True)

        median = 0 
        cum_sum = stations.weight.cumsum()
        for i in range(len(cum_sum)):
            if cum_sum.iloc[i] >= cum_sum.iloc[-1] / 2:
                median = i 
                break    
        if median == len(cum_sum)-1:
            median -= 1
            
        if isLa == True:    
            return Node(stations.iloc[median]['latitude']-0.00001,
                        None, False, None,
                        self.creat_tree(stations[:median+1],depth+1),
                        self.creat_tree(stations[median+1:],depth+1))
        elif isLo == True:
            return Node(stations.iloc[median]['longitude']-0.00001,
                        None, False, None,
                        self.creat_tree(stations[:median+1],depth+1),
                        self.creat_tree(stations[median+1:],depth+1))
        
```

```python
from __future__ import division
import pandas as pd
import random

regions = regions_num[10:]
count = 0

class QuadTree:
    def __init__(self, stations, depth, bounding_rect, maxIt):
        global count
        self.nw = self.ne = self.se = self.sw = None
        self.maxIt = maxIt
        self.depth = depth
        self.bounding_rect = bounding_rect
        cum_sum = stations.weight.cumsum()
        if cum_sum.iloc[-1] <= self.maxIt and depth >3:
            print '-----Leaf-----', cum_sum.iloc[-1], '----Stations----', len(stations)
            ML_station.loc[stations.index,'label']= regions[count]
            #DD.loc[stations.index,'label']= regions[self.depth]
            self.region = regions[count]
            count += 1
            self.items = stations.index
            self.leaf = True
            return
        self.leaf = False
        
        l, t, r, b = bounding_rect
        cx = self.cx = ( l + r ) * 0.5 
        cy = self.cy = ( t + b ) * 0.5 

        self.items = []
        nw_items = []
        ne_items = []
        se_items = []
        sw_items = []

        for item in stations.index:
            item_x = stations.loc[item, 'latitude']
            item_y = stations.loc[item, 'longitude']
            
            in_nw = item_x <= cx and item_y >  cy  
            in_ne = item_x >  cx and item_y >= cy
            in_sw = item_x <= cx and item_y <  cy  
            in_se = item_x >  cx and item_y <= cy
            
            if   in_nw: nw_items.append(item)
            elif in_ne: ne_items.append(item)
            elif in_sw: sw_items.append(item)
            elif in_se: se_items.append(item)

        if nw_items:
            self.nw = QuadTree(stations.loc[nw_items], depth+1, (l, t, cx, cy),maxIt)
        if ne_items:
            self.ne = QuadTree(stations.loc[ne_items], depth+1, (cx, t, r, cy),maxIt)
        if se_items:
            self.se = QuadTree(stations.loc[se_items], depth+1, (cx, cy, r, b),maxIt)
        if sw_items:
            self.sw = QuadTree(stations.loc[sw_items], depth+1, (l, cy, cx, b),maxIt)
```

```python
%%writefile pca_test.py
#!/usr/bin/python
import sys
sys.path.append('/usr/lib/python2.6/dist-packages')
import re,pickle,base64,zlib
from mrjob.job import MRJob
from sys import stderr
import pandas as pd
import numpy as np
import sklearn as sk
import gzip
import pickle

f = gzip.open('ML_station_hash.pkl.gz', 'rb' )
pickleFile = pickle.Unpickler(f)
hashtable = pickleFile.load()
f.close()

class PCA_weather(MRJob): 

    #This loads the lookup file into a field on the object
    def load_data(self):
        global hashtable
        f = gzip.open('ML_station_hash.pkl.gz', 'rb' )
        pickleFile = pickle.Unpickler(f)
        hashtable = pickleFile.load()
        f.close()
    
    def mapper_pre(self, _, line):
        global hashtable
        elements = line.split(',')
        if elements[1] == 'TMAX' or elements[1] == 'TMIN':
            region = hashtable.get(elements[0])
            if region != None:
                yield (elements[0], elements[2], region), (elements[1], elements[3:])
                
    def reducer_pre(self, station_pair, weathers):
        weather_max = []
        weather_min = []
        valid_count = 0
        data_count = 0
        for weather in weathers:
            if weather[0] == 'TMAX': weather_max = weather[1]
            if weather[0] == 'TMIN': weather_min = weather[1]
            data_count += 1
        if data_count == 2 and len(weather_max) == 365 and len(weather_min) == 365:
            for i in range(365):
                if weather_max[i] != '' and weather_min[i] != '':
                    valid_count += 1
            if valid_count >= 180:
                weather_data = weather_max + weather_min
                yield station_pair[2], weather_data
            
    def reducer_pca(self, region, matrixs):
        M = pd.DataFrame(matrixs)
        M[M == ''] = float('NaN')
        M = M.astype(float)
        M = M.transpose()
        (columns,rows)= np.shape(M)
        Mean = np.mean(M, axis=1).values
        C=np.zeros([columns,columns])  
        N=np.zeros([columns,columns])

        for i in range(rows):
            row = M.iloc[:,i] - Mean
            outer = np.outer(row,row)
            valid = np.isnan(outer) == False
            C[valid] = C[valid]+ outer[valid]
            N[valid] = N[valid]+ 1
            
        valid_outer = np.multiply(1 - np.isnan(N),N>0)
        cov = np.divide(C,N)
        cov = np.multiply(cov, valid_outer)
        U, D, V = np.linalg.svd(cov)
        cum_sum = np.cumsum(D[:])/np.sum(D)
        for i in range(len(cum_sum)):
            if cum_sum[i] >= 0.99:
                ind = i 
                break
        yield region, [ind, cum_sum.tolist()]
    
    def steps(self):
        return [self.mr(mapper = self.mapper_pre, reducer=self.reducer_pre),
                self.mr(reducer = self.reducer_pca)]
        
if __name__ == '__main__':
    PCA_weather.run()
```

```python
def is_adjacent(region1, region2):
    [b_1, r_1, t_1, l_1] = region1
    [b_2, r_2, t_2, l_2] = region2
    
    if l_1 == l_2 or r_1 == r_2:
        if t_1 == b_2 or b_1 == t_2:
            return True
        else:
            return False        
    if t_1 == t_2 or b_1 == b_2:
        if r_1 == l_2 or l_1 == r_2:
            return True
        else:
            return False
    if r_1 == l_2 or l_1 == r_2:
        if t_1 >= t_2 and b_1 <= b_2:
            return True
        elif t_2 >= t_1 and b_2 <= b_1:
            return True
        else:
            return False
    if t_1 == b_2 or b_1 == t_2:
        if l_1 >= l_2 and r_1 <= r_2:
            return True
        elif l_2 >= l_1 and r_2 <= r_1:
            return True
        else:
            return False
    
    return False
        
```

```python
def leafs_to_list(root, leaf_list):
    if root.leaf == True:
        leaf_list.append(root)
    else:
        if root.nw != None: leafs_to_list(root.nw, leaf_list)
        if root.ne != None: leafs_to_list(root.ne, leaf_list)
        if root.sw != None: leafs_to_list(root.sw, leaf_list)
        if root.se != None: leafs_to_list(root.se, leaf_list)
```

```python
leaf_num = len(leaf_list)
neighbours = []
for i in range(leaf_num):
    neighbours.append([])

for i in range(leaf_num): 
    for j in range(i+1, leaf_num):
        if is_adjacent(leaf_list[i].bounding_rect, leaf_list[j].bounding_rect):
            neighbours[i].append(j)
            neighbours[j].append(i)
            
```
