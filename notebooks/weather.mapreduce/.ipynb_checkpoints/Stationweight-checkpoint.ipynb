** Next, we will compute the valid measurements of each station all over the world. We will show that most of valid measurments are obtained in U.S stations suggesting that it's very reasonable to focus on the data analysis of U.S stations.**

** Mapreducer construciton for counting the number of valid measurement in each station. Again, a valid good measurement is defined as a year measurement including at least more than 50% days having both Tmax and Tmin measurements. "indicator = 1" if it's a valid measurement, otherwise "indicator = 0" **
> 1.mapper :  [ inputline ] ====> [ $key$ = [station,year], $value$ = Tmax_yeardata  $or$  Tmin_yeardata ]

> 2.reducer1:  [ $key$ = [station,year], $value$ = vector[Tmax_yeardata,Tmin_yeardata] ] ====> [ $key$ = station,$value$ = indicator ]

> 3.reducer2:   [$key$ = station, $value$ = [indicator,...,indicator] ] ====> [ $key$ = station, $value$ = sum([indicator,...,indicator]) ]

**After the Mapreducer, we will obtain the weight of each station**

```python
%%writefile valid_measurement.py
from __future__ import division
from mrjob.job import MRJob

class ValidMeasurement(MRJob):
    def mapper1(self, _, line):
        elements = line.split(',')
        if elements[1] == 'TMIN' or elements[1] == 'TMAX':
            yield (elements[0], elements[2]), elements[3:]

    def reducer1(self, pair, vectors):
        vectors = list(vectors)
        if len(vectors) < 2:
            yield pair[0], 0
        else:
            day_count = 0 
            for i in range(len(vectors[0])):
                if vectors[0][i] != '' and vectors[1][i] != '': 
                    day_count += 1
            if day_count / len(vectors[0]) < 0.5:
                yield pair[0], 0
            else:
                yield pair[0], 1
    
    def reducer2(self, station, is_valid):
        weight = sum(is_valid)
        if weight != 0:
            yield station, weight
    
    def steps(self):
        return [self.mr(mapper=self.mapper1, reducer=self.reducer1),
                self.mr(reducer=self.reducer2)]

if __name__ == '__main__':
    ValidMeasurement.run()
```

** Next, we will append the weight information from "stations_weight.txt" to the "station_info". We also carfully design the regular expression to include one special station name "BR00D6-0010" which contains a special character "-" **

```python
f = open('stations_weight.txt', 'rb')
d = []
for line in f.readlines():
    #"BR00D6-0010" has '-'
    data = re.compile(r"[\.\w-]+").findall(line)
    d.append(data)
header = ['station', 'weight']  
df = pd.DataFrame(d,columns=header)
df['weight'] = df['weight'].astype(float)
Djoined = df.join(station_info, on='station')
f.close()
```

** The codes in the following blocks will label three distinct regions in U.S: Mainland,Alaska,and Hawaii **

```python
def draw_screen_poly(plt, gid, west, east, south, north, m, color, text=True):
    def fix(x):
        if x >  80: return 80.0
        if x < -80: return -80.0
        return x    
    lons = [west, east, east, west]
    lats = map(fix, [south, south, north, north])
    x, y = m(lons, lats )
    xy = zip(x,y)    
    poly = Polygon( xy, facecolor=color, alpha=0.4, label = str(gid) )

    if text:
        # center    
        clon = (west + east) / 2
        clat = fix((north + south) / 2)
        cx,cy = m([clon],[clat])
        plt.text(cx[0], cy[0], str(gid), 
                 horizontalalignment='center', 
                 verticalalignment='center',
                 bbox=dict(facecolor='white', alpha=1),
                 size =15, fontweight='bold')  
       
    plt.gca().add_patch(poly)  
```
