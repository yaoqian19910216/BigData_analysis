## Sorting in memory

```python
import sys
from time import time
N=[];T1=[];T2=[]

for n in [10**i for i in range(3,7)]:
    t1=time()
    try:
        data=random.rand(n)
    except:
        print 'out of memory'
        break
    t2=time()
    sorted_data=sorted(data)
    t3=time()
    N.append(n)
    T1.append(t2-t1)
    T2.append(t3-t2)
    print 'size=',n,'allocation time=',t2-t1,'sorting time=',t3-t2
    sys.stdout.flush()
```

The run-time of sorting grows a little faster than linear, $O(n\log(n))$

Attempting to allocation an array of 10,000,000 entries  = 80MB crashes the kernel on a t3.micro instance  
Takes 11.2 seconds on my Apple Powerbook.

## Bucket sort for uniformly distributed numbers

```python
import struct, os
from array import array
def bucket_sort(filename,n=1000,buffers=100):
    """ sorts a file containing 8 byte floating point numbers into another file.
    filename: the name of the file whose content is to be sorted, output is filename+'.sorted'
    n: the size of the memory blocks used in each internal iteration'
    blocks: the number of buffer files used to store intermediate results
    """
    infile = open(filename,'rb')
    Buf=[[]]*buffers
    Buf_size=[0]*buffers
    A = array('d')
    t1=time()
    #Open "buffers" buffers for intermediate storage
    for j in range(buffers):
        Buf[j]=open(filename+'.buffer.'+str(j),'wb')
    count=0
    # split the input file into the buffers
    while True:
        s=infile.read(8*n)
        if not s:
            break
        A=array('d')  # clear array
        A.fromstring(s)
        if count%100==0: print count,'\r',
        count+=1
        for a in A:
            index=int(a*buffers)
            Buf[index].write(struct.pack('d',a))
            Buf_size[index]+=1
    t2=time() 
    # sort the content of each buffer
    outfile=open(filename+'.sorted','wb')
    for j in range(buffers):
        Buf[j].close()
    print
    for j in range(buffers):
        bufname=filename+'.buffer.'+str(j)
        print '\rsorting buffer',bufname,
        sys.stdout.flush()
        Buf[j]=open(bufname,'rb')
        data=array('d')
        data.fromfile(Buf[j],Buf_size[j])
        sorted_data=array('d',sorted(data))
        sorted_data.tofile(outfile)

    # Cleanup
    for j in range(buffers):
        Buf[j].close()
    outfile.close()
    t3=time()
    bufnames=filename+'.buffer*'
    !rm $bufnames
    print '\nsplitting time =',t2-t1,'sorting time =',t3-t2
    return t2-t1,t3-t2

bucket_sort('file.100000',buffers=100,n=1000)
```

## Run lengthy experiment

```python
# A script for running the bucket_sort routine using different parameters - takes many hours to complete...
filename='file.10000000'
!ls -l $filename
k=20 # repeates
results=[[]]*k
for i in range(k):
    results[i]={}
    for buffers in [10,20,50,100]:
        for n in [10,100,1000,10000]:
            print buffers,n
            results[i][(buffers,n)]=bucket_sort(filename,buffers=buffers,n=n)
```

## Save results in a pickle file

Pickle files provide a convenient way of storing complex data structures because it preserves the structure of the data structure.  
We say that pickle **serializes** the data structure.

When we run a lengthy experiment with many results it is a good idea to store the results in a large data structure that can be serialized to disk and later restored for detailed analysis. This is much more useful than just printing the results and later having to write code to parse the logs.

We see that increasing $n$ or $buffers$ generally results in faster execution.  
However there are some strange behaviours: $n=1000,buffers=50$ is faster than both $n=10000,buffers=50$ and $n=1000,buffers=100$,
and $n=100, buffers=100$ is better than $n=100, buffers=1000$  

My explanation is that while the general trend is for larger $n$ and larger $buffers$, there are values of these parameters that match the block sizes in the hardware and therefor cause less thrashing.
