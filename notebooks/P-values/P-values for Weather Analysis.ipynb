### NaN-tolerant averaging  ###
We compute the empirical covariance matrix in a way that tolerates NaN values.

<span style="color:red"> In the code below I remove all rows that have a nan in them. If you remve the command **M.dropna(...** then all rows are used. Can you get better results without removing the rows? </span>

```python
Null_Rows=pd.isnull(Dout.loc[:,1:365]).any(axis=1)
print len(Null_Rows),sum(Null_Rows)
```

```python
# demonstrating the use of the cell magic %%time, which measures the run-time of the cell.
M=Dout.loc[Null_Rows==False,1:365].transpose()
(columns,rows)=shape(M)
Mean=mean(M, axis=1).values   #Compute the means of just the full rows (no NaNs)

print (columns,rows), shape(Mean)
C=np.zeros([columns,columns])   # Sum
N=np.zeros([columns,columns])   # Counter of non-nan entries
```

```python
%%time
for i in range(rows):
    if i % 1000==0: 
        print i
    row=M.iloc[:,i]-Mean;
    outer=np.outer(row,row)
    valid=isnan(outer)==False
    C[valid]=C[valid]+outer[valid]  # update C with the valid location in outer
    N[valid]=N[valid]+1
valid_outer=np.multiply(1-isnan(N),N>0)
cov=np.divide(C,N)
```

```python
sqrt(sum(valid_outer==1))
```

```python
shape(cov)
```

```python
U,D,V=np.linalg.svd(cov)
```

```python
shape(U),shape(D),shape(V)
```

### Quantifying the statistical significance of the correlated eigen coefficient pairs ###  

Correlations are normalized so that they are always in the range $[-1,+1]$. The correlation between V3 and V2 is 0.177. Is this level of correlation statistically significant for a sample of the size we are using?

To answer the question we need a statistical test. The most common statistical test for strength of correlation is the pearson test. 

```python
from pydoc import help
from scipy.stats.stats import pearsonr,spearmanr
```

```python
help(pearsonr)
```

```python
def pearson(Data,select):
    corr_Table=pd.DataFrame(None,index=select,columns=select)
    p_Table=pd.DataFrame(None,index=select,columns=select)
    for e1 in select:
        for e2 in select:
            c,p=pearsonr(Data[e1].values,Data[e2].values)
            corr_Table.loc[e1,e2]=c
            p_Table.loc[e1,e2]=p
    return corr_Table,p_Table

corr_Table_pearson,p_Table_pearson=pearson(Djoined,selectV)
p_Table_pearson
```

We see that the correlation between V3 and V4 is highly significant. Other off-diagonal entries also show a significant level of correlation. There must be an explanation for this....

**The Mistake:**: our mistake is that while we constructed the PCA using only "full" rows, i.e. rows with no missing elements, we computed the correlation using **all** of the rows in Djoin.
To fix this what we should do is to recompute the correlation using only the full rows.

```python
print shape(Null_Rows),sum(Null_Rows),shape(Djoined)
Djoined_Full=Djoined[Null_Rows==False]
shape(Djoined_Full)
```

```python
corr_Table_pearson,p_Table_pearson=pearson(Djoined_Full,selectV)
p_Table_pearson
```

```python
corr_Table_pearson
```

### Spearman Test ###
So this seems to have closed the story. However, there is still one nagging question: the pearson test is based on the assumption that the joint distribution is normal, but that is clearly not true. How about using a statistical that does not require making this assumption.

Clearly, if a test makes fewer assumptions, the harder it is to fail the test. In other words, the p-values of the more liniant test should be higher than those of the more strict test.

```python
help(spearmanr)
```

```python
def spearman(Data,select):
    rho,pval=spearmanr(Data[select].values,axis=0)
    rho_Table=pd.DataFrame(rho,index=select,columns=select)
    p_Table=pd.DataFrame(pval,index=select,columns=select)
    return rho_Table,p_Table

```

```python
Rho_Table_1,p_Table_1=spearman(Djoined,selectV)
p_Table_1
```

```python
Rho_Table_2,p_Table_2=spearman(Djoined_Full,selectV)
p_Table_2
```

Strangely, it seems that the spearman test finds significant correlations even in the corrected table.
most of the p values have become much larger, indicating lower statistical significance. However the p-value of the correlation between V0 and V1 is still very low - no the order of $10^{-33}$.

This behaviour seems to contradict the definitions of the test. The pearson a high P-value from the pearson test indicates that the data has **high probability** of being generated from **a jointly normal distribution with no correlation between the components**. On the other hand a low P-value from spearman test indicates that the data has **low probability** of being generated from **an arbitrary distribution with no correlation between the components**.

This is very strange. The only explanation I can think of is that the implementation of the spearman test is incorrect. Checking that would require redoing the analysis, on the same data, using another platform such as **R** or **Matlab**.

### Permutation tests ###

```python
def permute_test(Data,select,iterations=100):
    D=pd.DataFrame(Data[select])
    Count=pd.DataFrame(0,index=select,columns=select)
    Corr_Table = D.corr() # correlation coefficients before permuting data
    for i in range(iterations):
        for col in select:
            D[col]=permutation(D[col])
        rand_corr = D.corr()  # correlation coefficients after permuting data
        better=rand_corr >= Corr_Table
        Count = Count + better
    return Count/iterations
        
permute_test(Djoined,selectV)    
```

### Reconstruction ###

```python
def plot_reconstructions(selection,rows=2,columns=7):
    Recon=Eig*Prod.transpose()+Mean[:,np.newaxis]
    plt.figure(figsize=(columns*3,rows*3),dpi=300)
    j=0;
    for i in selection:
        subplot(rows,columns,j); 
        j += 1; 
        if j>=rows*columns: break
        plot(Recon[:,i])
        plot(Djoined.ix[i,1:365]);
        title(Djoined.ix[i,'station']+' / '+str(Djoined.ix[i,'year']))
        xlim([0,365])
```

Observe in the reconstructions below that the bloue line fills in (extrapolation/interpolation) the places where the measurements are not available. It also reduces the fluctuations in the relative to the original line. Recall the we are using the k top eigenvectors which explain about 88% of the variance.

```python
plot_reconstructions(range(50,100),rows=7,columns=3)
```
