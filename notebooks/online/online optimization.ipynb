## Statistical Inference by minimizing a loss function ##
In the previous lessons we have learned about statistical inference problem that can be stated as minimization of some **loss function**. The loss function measures the degree of fit between the parameters of the statistical model, which we denote by $\theta$, and the data, which we denote by $X$.

#### Previous Examples ####
* **Mean:** Find the vector $\mu$ such that $\Phi_{\mu}(\mu,X) = \frac{1}{n} \sum_{i=1}^n \|\mu - x_i\|_2^2$ is minimized.
* **Largest Principle Component:** Find the unit length eigen-vector $v, \|v\|_2 = 1$ 
that minimized 
$$\Phi_{PC1}(v,\Sigma) \doteq  -v^\top \Sigma v$$
where $\Sigma \doteq \frac{1}{n} \sum_{i=1}^n (x_i-\mu) (x_i-\mu)^{\top}$.   
**Equivalently**, maximize the square length of the projection of the data on $v$, which corresponds to **minimizing** the loss function:
$$\Phi_{PC1}(v,\Sigma) = - \frac{1}{n} \sum_{i=1}^n v^\top (x_i-\mu) (x_i-\mu)^{\top} v 
= - \frac{1}{n} \sum_{i=1}^n (v^\top (x_i-\mu))^2$$
* **Linear Regression -- Ordinary Least Squares**  
$$\Phi_{OLS}(\vec{h},X,y) = \frac{1}{n} \sum_{i=1}^n \left( \vec{h}\cdot \vec{x}_i - y_i  \right)^2 $$
* ** Regularized Linear Linear -- Ridge Regression:**  
$$\Phi_{Ridge}(\vec{h},X,y) = \frac{1}{n} \sum_{i=1}^n \left(\left( \vec{h}\cdot \vec{x}_i - y_i \right)^2 +\lambda \|\vec{h}\|_2^2 \right)$$
* **Go to regression notebook** to see the role of regularization in batch optimization.

#### New Examples ####
* **Linear Discriminator** Similarly to the linear discriminator problem we are given a data matrix $X$, but unlike the regression problem the output $y$ is either $-1$ or $+1$. The loss function is the fraction of the coordinates in which  the prediction $\hat{y}_i$ and the label $y_i$ disagree. Formally:
$$\Phi_{Lin}(\vec{h},X,y) = \frac{1}{n} \sum_{i=1}^n \mathbb{1}\left[\mbox{sign}(\vec{h}\cdot \vec{x}_i) \neq y_i \right]$$
where $\mathbb{1}[cond]$ is the indicator function that is equal to 1 if the condition $cond$ holds and is 0 otherwise.
* **Logistic Regression** See [Logistic Regression](http://en.wikipedia.org/wiki/Logistic_regression)

## Gradient based model fitting ##

We have seen a few examples of loss functions. We will now look at some general schemes for minimizing loss.

Let $X$ be the data matrix, consisting of independent random vectors and let $\omega$ be the parameter vector.
Consider the following steps.

1. Define a loss function to be minimized. This function is usually in the form of a sum over the examples.
$$\omega^*(X) = \mbox{argmin}_\omega \Phi(\omega,X) = \mbox{argmin}_\omega \sum_{i=1}^n \phi(\omega,x_i)$$. Note that the loss function depends on a random variable: the sample $X$. As a result $\omega^*(X)$ is also a random variable. It is our hope that, as the size of the sample increases, the value of $\omega(X)$ is highly concentrated around some limit point $\omega^*_{\infty}$
1. **Batch algorithm:** find $\omega$ so that the gradient is zero $\bigtriangledown_\omega \Phi(\omega,X)=0$. When
his equation has a closed form solution we usually get an efficient optimization algorithm. However this algorithm typically requires all of $X$ to be in memory.
1. **Gradient Descent algorithm:** When there is no closed-form solution for the equation $\bigtriangledown_\omega \Phi(\omega,X)=0$. We can use a gradient descent method to search for the parameters with minimal loss. A gradient decent step is $\omega_{t+1}=\omega_t - \eta \bigtriangledown_\omega \Phi(\omega,X)$. A major disadvantage of the gradient descent method is that all instances in the dataset need to be processed before each update step. When we have a **lot** of data this might be prohibitively slow. In fact, we might be able to reach a good solution without reading all of the data (assuming that the data comes at a random order **HW: Why?**
1. **Online Algorithm:** based on reading into memory one example at a time and taking a small step in the direction of the gradient **wrt a single example: ** $$\omega_{t+1}=\omega_t - \eta \bigtriangledown_\omega \phi(\omega_t,x_i)$$
1. The learning rate $\eta$ defines a tradeoff between reducing the loss and keeping $\omega_{t+1}$ close to $\omega_t$. If we set the learning rate too high then the algorithm **forgets** too much of the solution-so-far after each update - will not converge. If we set the learning rate too low then the algorithm converges very slowly.  

## The error surface of ordinary least square ##

To understand the behaviour of stochastic gradient descent it is useful to look at dependence of the average loss function on the parameter vector $\vec{h}$. To make plotting possible we restrict ourselves to linear regression where the input dimension is one.

```python
import numpy as np
def predict(h,x):
    """ 
    predict a scalar y given x where y=h[0] + h[1]*x
    """
    return h[1]*x+h[0]*np.ones(len(x))
    
```

```python
x=np.arange(-10.0,10,0.1)
n=len(x)
h=np.array([0.0,10.0])
y=predict(h,x)+randn(n)*10
print shape(x),shape(y)
plot(x,y,'.')
ylim([-120,120])
grid()
```

```python
def calc_fit(x,y,h,plt=False):
    """ Calculate mean square error of h wrt x,y """
    ye=predict(h,x)
    if plt:
        plot(x,y,'r.')
        plot(x,ye)
    return mean(square(ye-y))
calc_fit(x,y,np.array([0,10]),plt=True)
```

```python
delta = 0.1
h0 = np.arange(h[0]-4.0, h[0]+4.0, delta)
h1 = np.arange(h[1]-4.0, h[1]+4.0, delta)  # h[1]-1.0, h[1]+1.0  is useful for 
                                           # a more detailed view (albeit with more bias)
H0, H1 = np.meshgrid(h0,h1)
E=np.zeros([len(h1),len(h0)])
for i0 in range(len(h0)):
    for i1 in range(len(h1)):
        E[i1,i0]=calc_fit(x,y,np.array([h0[i0],h1[i1]]))
# Create a simple contour plot with labels using default colors.  The
# inline argument to clabel will control whether the labels are draw
# over the line segments of the contour, removing the lines beneath
# the label
fig=plt.figure(1,figsize=(8,6),dpi=300);
CS = plt.contour(H0, H1, E,30);
plt.grid()
plt.clabel(CS, inline=1);
plt.xlabel('h[0]');
plt.ylabel('h[1]');
plt.title('Contour diagram of Mean square error as a function of h[0],h[1]');
```

Clearly, the quality of the approximation depends on $h_1$ more than it depends on $h_0$. The effect of this is also that the location of the minimum is much close to $h_1=10$ than to $h_0=0$

## LMS - Least Mean Square (Widrow-Hoff)##
The LMS algorithm is a very simple one, read more about it [here](http://en.wikipedia.org/wiki/Least_mean_squares_filter).

LMS is a stochastic gradient descent method for the ordinary least-squares loss function:
$$\Phi_{OLS}(\vec{h},X,y) = \frac{1}{n} \sum_{i=1}^n \left( \vec{h}\cdot \vec{x}_i - y_i  \right)^2 $$
The gradient of a single term with respect to $\vec{h}$ is
$$ \bigtriangledown_{\vec{h}} \left( \vec{h}\cdot \vec{x}_i - y_i  \right)^2 = 2(\vec{h}\cdot\vec{x}_i-y_i)\vec{x}$$
To keep the discussion intuitive and plottable, we will restrict our view to two dimensional case.

```python
def LMS_learn(x,y,mu=0.001):
    #ha is our approximation of h
    ha = random.randn(2)
    t=1
    while True:
        for i in range(len(x)):
            yield ha   #this function will be used as a *generator* in a for loop
            ya= ha[0]+ha[1]*x[i]
            err= y[i]-ya
            ha=ha+mu*err*np.array([1,x[i]])
            t+=1
```

## RLS  - Recursive Least Squares ##

The recursive least square method is, in most cases, better than LMS. Instead of considering only the gradient of the loss function at the current point it uses a second-order estimate of the shape of the potential function around the current point. Using this estimate, denoted $P$ it "normalizes" the error surface so that it close to spherical. The gradient in the renormalized space points in the direction of the minimum and the convergence is faster.  
We say that RLS is a **second order** method because it depends on the first and second derivatives -- on the Gradient and on the Hessian. [You can read more about RLS in this wikipedia page](http://en.wikipedia.org/wiki/Recursive_least_squares)

Here is an implementation of the algorithm described on wikipedia.

```python
def RLS_learn(x,y,Lambda=0.99,delta=1):
    #ha is our approximation of h
    ha = np.zeros(2)
    P=delta*np.eye(2)
    t=1
    while True:
        for i in range(len(x)):
            xv=np.array([1,x[i]])
            ya= np.dot(xv,ha)
            alpha= y[i]-ya
            g=np.dot(P,xv)/(Lambda+np.dot(xv.transpose(),np.dot(P,xv)))
            P=(P-np.dot(np.outer(g,xv),P))/Lambda
            ha=ha+alpha*g
            yield ha,P,g,alpha,ya,xv
            t+=1
```

## The perceptron algorithm ##

We cannot minimize the linear discriminator loss function by using gradient descent. That is because the gradient of the
loss function is either undefined or zero.
$$\Phi_{Lin}(\vec{h},X,y) = \frac{1}{n} \sum_{i=1}^n \mathbb{1}\left[\mbox{sign}(\vec{h}\cdot \vec{x}_i) \neq y_i \right]$$

Instead we can minimize a related loss function, the **hinge** loss function:
$$\Phi_{Lin}(\vec{h},X,y) = \frac{1}{n} \sum_{i=1}^n \mbox{Hinge}\left[(\vec{h}\cdot \vec{x}_i) y_i\right]$$
where 
$$\mbox{Hinge}(x)=\begin{cases}-x & x \leq 0\\0 & x > 0\end{cases} 
$$

**Excercise:** implement the stochastic gradient descent algorithm for the hinge loss, also known as the **Perceptron algorithm**

###making the tradeoff between adaptivity and stabilty explicit###
We can define the update rule by making this tradeoff explicit (here we are talking about the linear regression problem, similar statements can be made for other loss functions)
$$\omega_{t+1} = \mbox{argmin}_{\omega} \left(\|\omega-\omega_t\|_2^2 + \eta (y_t-\omega\cdot x_t)^2 \right) = 
\omega_t - \eta\left(\omega_{t+1}\cdot x_t -y_t \right) x_t \approx \omega_t - \eta\left(\omega_{t}\cdot x_t -y_t \right) x_t$$
The approximation in the last step reflects the fact that in RLS we use the gradient at $\omega_t$, not at $\omega_{t+1}$.
Note that $(y_t-\omega\cdot x_t)^2$ is a loss term and $\|\omega-\omega_t\|_2^2$ is a regularization term. These terms play a similar role in Ridge regression - where the algorithm minimizes a combination of the loss and of the **complexity** or length of $\omega$.
1. One interesting direction that comes out of the approach is that the squared euclidean norm is not the only measure of change that can be used in this context. Another interesting measure is the Kullback divergence which measures the distance between distribution vectors. (In other words, vectors all of whose components are non-negative and sum to one). If $\vec{u}$ and $\vec{v}$ are both $n$ dimensional probability vectors then the KL-divergence is defined as:
$$D_{kl}(\vec{u}||\vec{v}) \doteq \sum_{i=1}^n u_i \ln \frac{u_i}{v_i}$$
Using $D_{kl}$ instead of the square euclidean norm in step 5. above we get:
$$\omega_{t+1} = \mbox{argmin}_{\omega} \left(\sum_{i=1}^n \omega_i \ln \frac{\omega_i}{\omega_{t,i}} + \eta (y_t-\omega\cdot x_t)^2 \right) = \frac{1}{Z} \omega_{t,i} \exp \left[ -\eta(\omega_{t+1}\cdot x_t-y_t)\;x_{t,i}  \right]
\approx \frac{1}{Z'} \omega_{t,i} \exp \left[ -\eta(\omega_{t}\cdot x_t-y_t)\;x_{t,i}  \right]$$
1. We will not pursue the mathematical analysis of this kind of algorithm further at this point. The interested student might want to have a look at ["Prediction, Learning and Games" by Cesa Binachi and Lugosi](http://www.ii.uni.wroc.pl/~lukstafi/pmwiki/uploads/AGT/Prediction_Learning_and_Games.pdf)

####Batch versus Online ####
**Batch** methods for solving these problems amount to equating the gradient of the loss function to zero. 
$$\frac{\partial}{\partial \theta_i} \Phi(\theta,X)=0$$
This is a good approach when data $X$ is small enough to fit in memory. When we have more data than can fit in memory we need to use something else.

One powerful solution is **regularized stocahstic gradient descent**. The power of this approach is in that the estimated parameter $\hat{\theta}$ is updated sequentially and each update depends only on a single example. As a result, implementing these examples requires only enough memory to store a single example. The pattern of operation in which the algorithm processes all examples in a sequence in order is called a **streaming algorithm**. We will get back to streaming algorithms later in the course.
